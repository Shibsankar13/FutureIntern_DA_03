```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/kaggle/input/iris/Iris.csv'
df = pd.read_csv(file_path)

# Create a figure to hold multiple plots for the numerical features
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Create histograms for the numerical features
df['SepalLengthCm'].hist(ax=axes[0, 0], bins=15, color='skyblue', edgecolor='black')
axes[0, 0].set_title('Distribution of Sepal Length')
axes[0, 0].set_xlabel('Sepal Length (cm)')
axes[0, 0].set_ylabel('Frequency')

df['SepalWidthCm'].hist(ax=axes[0, 1], bins=15, color='lightgreen', edgecolor='black')
axes[0, 1].set_title('Distribution of Sepal Width')
axes[0, 1].set_xlabel('Sepal Width (cm)')
axes[0, 1].set_ylabel('Frequency')

df['PetalLengthCm'].hist(ax=axes[1, 0], bins=15, color='lightcoral', edgecolor='black')
axes[1, 0].set_title('Distribution of Petal Length')
axes[1, 0].set_xlabel('Petal Length (cm)')
axes[1, 0].set_ylabel('Frequency')

df['PetalWidthCm'].hist(ax=axes[1, 1], bins=15, color='gold', edgecolor='black')
axes[1, 1].set_title('Distribution of Petal Width')
axes[1, 1].set_xlabel('Petal Width (cm)')
axes[1, 1].set_ylabel('Frequency')

# Adjust layout for the histograms
plt.tight_layout()
plt.show()

```

```python
# Bar chart for Species
plt.figure(figsize=(8, 6))
df['Species'].value_counts().plot(kind='bar', color='lightblue', edgecolor='black')
plt.title('Distribution of Iris Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
